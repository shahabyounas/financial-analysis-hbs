```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_excel('shell.xlsx', header=0, names=['Ratios','FY Dec-18', 'FY Dec-19','FY Dec-20','FY Dec-21','FY Dec-22'])

dfc = df.dropna()

# df.info()

# df.describe()

col = 'Ratios'

dfc.loc[:, col] = df[col].str.replace(' ', '_').str.lower()

# dfc.set_index('Ratio', inplace=True)

gpm = dfc[dfc['Ratios'] == 'gross_profit_margin']

# dfc.iloc[5:15].iloc(0)

# dfc['Ratios'] 

gpm


# Create a DataFrame with the provided data
data = {
    'Year': [2018, 2019, 2020, 2021, 2022],
    'Shell': [17.25, 18.98, 20.88, 15.19, 25.52],
    'BP': [14.84, 16.29, 13.90, 24.71, 29.72],
    'Harbour Energy': [-0.18, 30.54, 25.04, 68.02, 47.22],
    'Avg': [0.10637, 21.94, 19.94, 35.97, 34.15]
}

df = pd.DataFrame(data)

# Plot the line chart
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Shell'], marker='o', label='Shell')
plt.plot(df['Year'], df['BP'], marker='o', label='BP')
plt.plot(df['Year'], df['Harbour Energy'], marker='o', label='Harbour Energy')
plt.plot(df['Year'], df['Avg'], marker='o', linestyle='--', label='Industry Avg')

plt.title('Gross Profit Margin Comparison')
plt.xlabel('Years')
plt.ylabel('Gross Profit Margin (%)')
plt.legend()
plt.grid(True)

plt.savefig('gross_profit_margin_comparison.png')
plt.show()



```
